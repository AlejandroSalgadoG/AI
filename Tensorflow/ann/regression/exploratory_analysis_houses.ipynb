```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("../materials/DATA/kc_house_data.csv")
data.head()
```

```python
numeric_cols = [
    'price',
    'sqft_living',
    'sqft_lot',
    'sqft_above',
    'sqft_basement',
    'sqft_living15', 
    'sqft_lot15'
]
numeric_data = data[numeric_cols]
numeric_data.head()
```

```python
numeric_data.describe().T
```

```python
sns.heatmap(numeric_data.corr(), annot=True, cmap='BrBG')
```

```python
sns.displot(data['price'])
```

```python
common_price_data = data[data["price"] < 3e6]
print(common_price_data.shape)
sns.displot(common_price_data['price'])
```

```python
sns.scatterplot(x='price',y='sqft_living',data=common_price_data)
```

```python
sns.countplot(common_price_data, x='bedrooms')
```

```python
sns.boxplot(x='bedrooms',y='price',data=common_price_data)
```

```python
sns.scatterplot(x='long', y='lat',data=common_price_data, hue='price')
```
